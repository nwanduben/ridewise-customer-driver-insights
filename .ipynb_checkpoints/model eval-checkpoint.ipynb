```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import roc_auc_score, precision_score, mean_absolute_error, r2_score
from sklearn.metrics import roc_curve, precision_recall_curve
from imblearn.over_sampling import SMOTE
import xgboost as xgb
import shap
import matplotlib.pyplot as plt
from datetime import datetime

```

```python
df = pd.read_csv('data/processed/riders_enriched.csv',parse_dates=['signup_date'])
```

```python
# Create churn column by rounding churn_prob
df['churn'] = df['churn_prob'].round().astype(int)
```

```python
df.head()
```

```python
 # Analyze churn distribution by city and loyalty_status (before dropping)
churn_by_city = df.groupby('city')['churn'].mean()
churn_by_loyalty = df.groupby('loyalty_status')['churn'].mean()

print("Churn Rate by City:\n", churn_by_city)
print("\nChurn Rate by Loyalty Status:\n", churn_by_loyalty)

interesting_fact = "Users in {} have the highest churn rate ({:.2%}), and {} loyalty users have the highest churn rate ({:.2%}).".format(
    churn_by_city.idxmax(), churn_by_city.max(),
    churn_by_loyalty.idxmax(), churn_by_loyalty.max()
)
interesting_fact
```

```python
df.dtypes
```

```python
# Drop non-numeric/ID columns
columns_to_drop = ['user_id', 'signup_date', 'loyalty_status', 'city', 'referred_by']
df_numeric = df.drop(columns=columns_to_drop)
df_numeric.head()
```

```python
# Check class distribution
print("\nChurn Class Distribution:\n", df_numeric['churn'].value_counts())
```

```python
# Handle missing values by imputing with mean
df_numeric = df_numeric.fillna(df_numeric.mean())
df_numeric.head()
```

```python
# Define features and target
X = df_numeric.drop(columns=['churn', 'churn_prob'])
y = df_numeric['churn']

X.shape,y.shape
```

```python
# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)
```

```python
train_df =  (df['signup_date'].dt.year.isin([2023, 2024]))
test_df  =  (df['signup_date'].dt.year == 2025)

train_df.head()
```

```python
X_train = X_scaled[train_df]
y_train = y[train_df]
X_val = X_scaled[test_df]
y_val = y[test_df]
```

```python
X_train.shape, X_val.shape, y_train.shape, y_val.shape
```

```python
# Apply SMOTE to training data
smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)
```

```python
# Define expanded hyperparameter grid
log_reg_params = {
    'C': np.logspace(-3, 2, 6),  
    'penalty': ['l1', 'l2', 'elasticnet'],  
    'solver': ['liblinear', 'saga'],  
    'class_weight': ['balanced', None],  
    'l1_ratio': [0, 0.5, 1],  
    'tol': [1e-4, 1e-3]  
}

# Initialize LogisticRegression with fixed random_state
log_reg = GridSearchCV(
    LogisticRegression(random_state=42, max_iter=2000),  
    log_reg_params,
    cv=5,
    scoring='roc_auc',
    n_jobs=-1 
)

# Fit the model
log_reg.fit(X_train_smote, y_train_smote)

# Get the best estimator and parameters
best_log_reg = log_reg.best_estimator_
print("\nBest Logistic Regression Parameters:", log_reg.best_params_)
print("Best ROC AUC Score:", log_reg.best_score_)
```

```python
# Hyperparameter tuning for XGBoost
xgb_params = {
    'learning_rate': [0.01, 0.1, 0.3],
    'max_depth': [3, 5, 7],
    'n_estimators': [100, 200],
    'scale_pos_weight': [1, len(y_train_smote[y_train_smote == 0]) / len(y_train_smote[y_train_smote == 1])]
}
xgb_model = GridSearchCV(xgb.XGBClassifier(random_state=42, eval_metric='logloss'), xgb_params, cv=5, scoring='roc_auc')
xgb_model.fit(X_train_smote, y_train_smote)
best_xgb = xgb_model.best_estimator_
print("Best XGBoost Parameters:", xgb_model.best_params_)
```

```python
# Evaluate models
models = {'Logistic Regression': log_reg, 'XGBoost': xgb_model}
metrics = {}
roc_data = {}
pr_data = {}

for name, model in models.items():
    # Predictions
    y_pred = model.predict(X_val)
    y_pred_proba = model.predict_proba(X_val)[:, 1]
    
    # Metrics
    auc = roc_auc_score(y_val, y_pred_proba)
    precision = precision_score(y_val, y_pred)
    mae = mean_absolute_error(y_val, y_pred)
    r2 = r2_score(y_val, y_pred)
    metrics[name] = {'AUC': auc, 'Precision': precision, 'MAE': mae, 'R2': r2}
    
    # ROC Curve
    fpr, tpr, _ = roc_curve(y_val, y_pred_proba)
    roc_data[name] = pd.DataFrame({'FPR': fpr, 'TPR': tpr})
    
    # PR Curve
    precision_vals, recall, _ = precision_recall_curve(y_val, y_pred_proba)
    pr_data[name] = pd.DataFrame({'Precision': precision_vals, 'Recall': recall})

# Print metrics
print("\nModel Performance Metrics:")
for name, metric in metrics.items():
    print(f"\n{name}:")
    print(f"AUC: {metric['AUC']:.3f} {'✅' if metric['AUC'] > 0.85 else '❌'}")
    print(f"Precision: {metric['Precision']:.3f} {'✅' if metric['Precision'] > 0.8 else '❌'}")
    print(f"MAE: {metric['MAE']:.3f}")
    print(f"R2: {metric['R2']:.3f}")
```

```python
%matplotlib inline 
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
for name in pr_data:
    plt.plot(pr_data[name]['Recall'], pr_data[name]['Precision'], label=name)
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curves')
plt.legend()
plt.grid(True)
plt.savefig('pr_curves_improved.png')  # Save the plot
plt.show()  # Display the plot in the notebook
plt.close()  # Close the figure to free memory
```
